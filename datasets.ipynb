# Importation des bases de données pour le projet depuis data.gouv

#### Base disparité des niveaux de vie

#### Base jeunes non diplomés de 25-34 ans (2021 - MaJ 2024)

#### Intensité de la pauvreté (2021 - MaJ 2024)

```python
# URL directe du fichier
url = "https://www.data.gouv.fr/api/1/datasets/r/2ee6cff1-efeb-4739-bde3-d1778bea8b86"
# Lecture avec encodage adapté
df4 = pd.read_excel(url, engine='openpyxl', header = 5, nrows=101, usecols=[0,1,2]).rename(columns = {"Unnamed: 1": "Dep",
"Intensité de la pauvreté (1)": "intens_pauvr"}).drop(columns=["Départements"]).dropna()
# Vérification
df4
```

#### Taux de chomage (2019,2021,2024)

```python
# URL directe du fichier
url = "https://www.data.gouv.fr/api/1/datasets/r/500d1954-9849-479b-9aac-85435c6d422f"
# Lecture avec encodage adapté
df5 = pd.read_excel(url, engine='openpyxl', header = 5, usecols=[0,1,7,12],  #[0,1,7,10,12]
 nrows=101, sheet_name="Moyenne annuelle").rename(columns = {"Unnamed: 1": "Dep",
2019: "chom19", 2024: "chom24"}).drop(columns=["Département"]).dropna()
# Vérification
df5
```

#### Demandeurs d'emploi (Part de la population active 15-64 ans)


```python
# URL directe du fichier
url = "https://www.data.gouv.fr/api/1/datasets/r/87b12173-d9fb-462f-b216-370439ac6d69"
# Lecture avec encodage adapté
df6 = pd.read_excel(url, engine='openpyxl', header = 4, usecols=[0,1,8],
 nrows=101, sheet_name="Données").rename(columns = {"Unnamed: 1": "Dep",
"Part des demandeurs d’emploi dans la population en âge de travailler (15 -64 ans) (3)": "Dem_empl15_64"}).drop(columns=["Départements"]).dropna()
# Vérification
df6
```

#### Pauvreté monétaire (2021 - MaJ 2024)

```python
# URL directe du fichier
url = "https://www.data.gouv.fr/api/1/datasets/r/ad9653b3-0ee8-4286-b94f-b86522c25c61"
# Lecture avec encodage adapté
df7 = pd.read_excel(url, engine='openpyxl', header = 4, usecols=[0,1,2],
 nrows=101).rename(columns = {"Unnamed: 1": "Dep",
"Taux de pauvreté (1)": "tauxpauvrete"}).drop(columns=["Départements"]).dropna()
# Vérification
df7
```

### Fusion des différents tableaux pour avoir le df final

```python
from functools import reduce
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7]

# Fusion progressive sur la colonne "Dep"
df= reduce(lambda left, right: pd.merge(left, right, on="Dep", how="inner"), dfs)
df.head()
```

```python
# On sauvegarde le fichier dans le repo github
df.to_csv("df_regressions.csv", index=False)
```
